Use the Big-Mart Data for some plotting exercises

## Exercise 1
Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: 

* X label name = Month Number
* Y label name = Total profit


```python
#labels the x and y axis
plt.xlabel('Month number')
plt.ylabel('Total profit')
#access data total_profit
plt.plot(data.total_profit)
```

## Exercise 2
Get Total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

* Line Style dotted and Line color should be red
* Show legend at the lower right location.
* X label name = Month Number
* Y label name = Sold units number
* Add a circle marker.
* Line marker color as read
* Line width should be 3


```python
#dotted red line
plt.plot(data.total_profit, 'r:', linewidth=3.00)

#legend in the lower right corner
plt.legend("curve1", loc=4)

#X label name = Month number
plt.xlabel('Month number')

#Y label name = Sold units number
plt.ylabel('Sold Units number')

#circle marker color = red, linewidth = 3          not sure about this one
plt.plot(data.total_profit, color="red",lw=3, ls=':',  marker='o')
```

## Exercise 3
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).

```python
#labels the x and y axis
plt.xlabel('Month number')
plt.ylabel('Sold Units number')
#displays all the plot diagramms and labels them for the legend
plt.plot(data.facecream, label = "facecream")
plt.plot(data.facewash, label = "facewash")
plt.plot(data.toothpaste, label = "toothpaste")
plt.plot(data.bathingsoap, label = "bathingsoap")
plt.plot(data.shampoo, label = "shampoo")
plt.plot(data.moisturizer, label = "moisturizer")
#legend in the top right corner
plt.legend(loc = 2)
```

## Exercise 4
Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot.

```python
#shows the scatterplot with month number as x and sold units as y
plt.scatter(data.month_number, data.toothpaste)
#shows the grid
plt.grid(True)
```

## Exercise 5
Read face cream and facewash product sales data and show it using the bar chart
Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
#make a bar for the facecream and align it on the left edge
plt.bar(data.month_number,data.facecream, align="edge", width = 0.4)
#make a bar for the facewash and align it to the right edge; possible because of the negative value of width
plt.bar(data.month_number,data.facewash, align="edge", width=-0.4)
```
